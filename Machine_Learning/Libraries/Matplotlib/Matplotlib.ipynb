```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[2,3,4],"b+-",markersize=10,color="g",antialiased=True,linewidth=11,dashes=[2,3])
plt.plot([4,5,6,],linestyle="-")
# (colour marker line combination)
# There are many line style and marker. For that just google it
# linewidth -increase the width of the line 
# antialiased-helps to smothen the line 
# color-helps to change the colour of the graph
# dashes-helps to convert graph lines to dashes
plt.title("Plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# zorder-helps to move backward line to front and viceversa
plt.plot([0,0],[-1,1],linewidth=11,zorder=2)
plt.plot([-1,1],[-1,1],linewidth=11,zorder=1)
plt.plot([-1,1],[1,-1],linewidth=11,zorder=3)
```

```python
plt.plot([4,5,6,],linestyle="-")
plt.title("Plot")
plt.xlabel("X-axis",{"size":20,"color":"blue"},labelpad=10)
# labelpad helps to adjust position of x label
plt.ylabel("y-axis")
plt.show()
```
